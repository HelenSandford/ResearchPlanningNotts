Harvest a test set of data to csv to review fields available

```python
import requests
import csv
```

```python
# Fetch data (n.b. min fetchSize defaults to 25)
response = requests.get(
    "https://gtr.ukri.org/api/projects?page=1&fetchSize=10",
    headers={"Accept": "application/json"}
)
projects = response.json()["projectsBean"]["projects"]
```

```python
# Explore data to see fields available
print(projects[0].keys())
projects[0]
```

```python
# Save to csv
with open("gtr_projects.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=projects[0].keys())
    writer.writeheader()
    writer.writerows(projects)

print(f"Saved {len(projects)} projects to gtr_projects.csv")
```
